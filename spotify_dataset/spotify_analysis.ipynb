Import Data and Modules

```python
import pandas as pd
import seaborn as sns

data = pd.read_csv('spotify_dataset.csv')
```

Clean data

```python
data = data.replace(' ', None)
data = data.astype({'Popularity': 'int64',
                    'Danceability': 'float',
                    'Energy': 'float',
                    'Loudness': 'float',
                    'Speechiness': 'float',
                    'Acousticness': 'float', 
                    'Liveness': 'float', 
                    'Tempo': 'float', 
                    'Duration (ms)': 'int64',
                    'Valence': 'float'})
```

```python
correlations = data.corr()
ax = sns.heatmap(correlations)
```
